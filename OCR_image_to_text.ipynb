# OCR.space API 

Попробуем другой вариант с использованием API. Например, OCR.space - бесплатный сервис распознавания текстов с изображений

```python
import io
import json
import cv2
import numpy as np
import requests
from google.colab.patches import cv2_imshow

img = cv2.imread("Your image file")
height, width, _ = img.shape
```

```python
cv2_imshow(img)
```

```python
url_api = "https://api.ocr.space/parse/image"
_, compressedimage = cv2.imencode(".jpg", img, [1, 90])
file_bytes = io.BytesIO(compressedimage)
```

```python
result = requests.post(url_api,
              files = {"1.png": file_bytes},
              data = {"apikey": "YOUR API",
                      "language": "rus"})
```

```python
result
```

```python
result = result.content.decode()
result = json.loads(result)
```

```python
parsed_results = result.get("ParsedResults")[0]
text_detected = parsed_results.get("ParsedText")
print(text_detected)
```

Как видно с результатов, данный сервис справляется чуть лучше и в большинстве случаев правильно считывает чуть лучше чем tesseract-ocr, по крайней мере фамилию и имя выдает правильно.
